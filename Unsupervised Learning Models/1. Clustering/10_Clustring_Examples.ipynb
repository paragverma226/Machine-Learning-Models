```python
# Step 1: Importing the important libraries
"""
1. NumPy.where(): function returns the indices of elements in an input array where the given condition is satisfied.
2. NumPy.unique(): This numpy module of Python provides a function for finding unique elements in a numpy array. 
3. make_classification: Generate a random n-class classification problem.
4. Matplotlib: is a cross-platform, data visualization and graphical plotting library and its numerical extension NumPy.
"""

from numpy import where, unique
from sklearn.datasets import make_classification
import matplotlib.pyplot as plt
```

```python
# define dataset
X, y = make_classification(n_samples=1000, n_features=2, n_informative=2, n_redundant=0, n_clusters_per_class=1, random_state=4)
```

```python
X,y
```

```python
# create scatter plot for samples from each class
for class_value in range(2):
    row_ix = where(y == class_value) # get row indexes for samples with this class
    plt.scatter(X[row_ix, 0], X[row_ix, 1]) # create scatter of these samples
# show the plot
plt.show()
```

```python
# Algorithm 1: 
# Affinity propagation clustering: is a clustering algorithm based on the concept of "message passing" between data points.
from sklearn.cluster import AffinityPropagation
# Define the model
model = AffinityPropagation(damping=0.9)
# fit the model
model.fit(X)
# Assign a cluster to each example
yhat = model.predict(X)
# retrieve unique clusters
clusters = unique(yhat)

# create scatter plot for samples from each cluster
for cluster in clusters:
    row_ix = where(yhat == cluster) # get row indexes for samples with this cluster
    plt.scatter(X[row_ix, 0], X[row_ix, 1]) # create scatter of these samples
# show the plot
plt.title('Affinity propagation clustering')
plt.show()
```

```python
# Algorithm 2: 
# Agglomerative clustering: This is a "bottom-up" approach: each observation starts in its own cluster, and pairs of clusters are merged as one moves up the hierarchy.
from sklearn.cluster import AgglomerativeClustering
# Define the model
model = AgglomerativeClustering(n_clusters=2)
# fit model and predict clusters
yhat = model.fit_predict(X)
# retrieve unique clusters
clusters = unique(yhat)
# create scatter plot for samples from each cluster
for cluster in clusters:
    row_ix = where(yhat == cluster) # get row indexes for samples with this cluster
    plt.scatter(X[row_ix, 0], X[row_ix, 1]) # create scatter of these samples

# show the plot
plt.title('Agglomerative clustering')
plt.show()
```

```python
# Algorithm 3: 
# BIRCH (Balanced Iterative Reducing and Clustering using Hierarchies) clustering: is an unsupervised data mining algorithm used to perform hierarchical clustering over particularly large data-sets.
from sklearn.cluster import Birch
# Define the model
model = Birch(threshold=0.01, n_clusters=2)
# fit the model
model.fit(X)
# assign a cluster to each example
yhat = model.predict(X)
# retrieve unique clusters
clusters = unique(yhat)
# create scatter plot for samples from each cluster

for cluster in clusters:
    row_ix = where(yhat == cluster) # get row indexes for samples with this cluster
    plt.scatter(X[row_ix, 0], X[row_ix, 1]) # create scatter of these samples
    
# show the plot
plt.title('BIRCH clustering')
plt.show()
```

```python
# Algorithm 4: 
# Density-Based Spatial Clustering of Applications with Noise (DBSCAN) clustering: It can discover clusters of different shapes and sizes from a large amount of data, which is containing noise and outliers.
from sklearn.cluster import DBSCAN
# Define the model
model = DBSCAN(eps=0.30, min_samples=9)
# fit model and predict clusters
yhat = model.fit_predict(X)
# retrieve unique clusters
clusters = unique(yhat)
# create scatter plot for samples from each cluster

for cluster in clusters:
    row_ix = where(yhat == cluster) # get row indexes for samples with this cluster
    plt.scatter(X[row_ix, 0], X[row_ix, 1]) # create scatter of these samples
# show the plot
plt.title('DBSCAN clustering')
plt.show()
```

```python
# Algorithm 5:
# k-means clustering: is a method of vector quantization, originally from signal processing, 
# It's aims to partition n observations into k clusters in which each observation belongs to the cluster with the nearest mean (cluster centers or cluster centroid), serving as a prototype of the cluster.
from sklearn.cluster import KMeans
# Define the model
model = KMeans(n_clusters=2)
# fit the model
model.fit(X)
# assign a cluster to each example
yhat = model.predict(X)
# retrieve unique clusters
clusters = unique(yhat)
# create scatter plot for samples from each cluster
for cluster in clusters:
    row_ix = where(yhat == cluster) # get row indexes for samples with this cluster
    plt.scatter(X[row_ix, 0], X[row_ix, 1]) # create scatter of these samples

# show the plot
plt.title('k-means clustering')
plt.show()
```

```python
# Algorithm 6: 
# mini-batch k-means clustering
from sklearn.cluster import MiniBatchKMeans
# Define the model
model = MiniBatchKMeans(n_clusters=2)
# fit the model
model.fit(X)
# assign a cluster to each example
yhat = model.predict(X)
# retrieve unique clusters
clusters = unique(yhat)

# create scatter plot for samples from each cluster
for cluster in clusters:
    row_ix = where(yhat == cluster) # get row indexes for samples with this cluster
    plt.scatter(X[row_ix, 0], X[row_ix, 1]) # create scatter of these samples

# show the plot
plt.title('mini-batch k-means clustering')
plt.show()
```

```python
# Algorithm 7: 
# mean shift clustering
from sklearn.cluster import MeanShift
# Define the model
model = MeanShift()
# fit model and predict clusters
yhat = model.fit_predict(X)
# retrieve unique clusters
clusters = unique(yhat)

# create scatter plot for samples from each cluster
for cluster in clusters:
    row_ix = where(yhat == cluster) # get row indexes for samples with this cluster
    plt.scatter(X[row_ix, 0], X[row_ix, 1]) # create scatter of these samples

# show the plot
plt.title('mean shift clustering')
plt.show()
```

```python
# Algorithm 8: 
# Ordering points to identify the clustering structure (OPTICS) clustring: is an algorithm for finding density-based clusters in spatial data.
from sklearn.cluster import OPTICS
# define the model
model = OPTICS(eps=0.8, min_samples=10)
# fit model and predict clusters
yhat = model.fit_predict(X)
# retrieve unique clusters
clusters = unique(yhat)

# create scatter plot for samples from each cluster
for cluster in clusters:
    row_ix = where(yhat == cluster) # get row indexes for samples with this cluster
    plt.scatter(X[row_ix, 0], X[row_ix, 1]) # create scatter of these samples

# show the plot
plt.title('OPTICS clustering')
plt.show()
```

```python
# Algorithm 9: 
# Spectral clustering: techniques make use of the spectrum (eigenvalues) of the similarity matrix of the data to perform dimensionality reduction before clustering in fewer dimensions. 
from sklearn.cluster import SpectralClustering
# Define the model
model = SpectralClustering(n_clusters=2)
# fit model and predict clusters
yhat = model.fit_predict(X)
# retrieve unique clusters
clusters = unique(yhat)

# create scatter plot for samples from each cluster
for cluster in clusters:
    row_ix = where(yhat == cluster) # get row indexes for samples with this cluster
    plt.scatter(X[row_ix, 0], X[row_ix, 1]) # create scatter of these samples

# show the plot
plt.title('Spectral clustering')
plt.show()
```

```python
# Algorithm 10: 
# Gaussian Mixture clustering: model summarizes a multivariate probability density function with a mixture of Gaussian probability distributions as its name suggests.
from sklearn.mixture import GaussianMixture
# Define the model
model = GaussianMixture(n_components=2)
# fit the model
model.fit(X)
# assign a cluster to each example
yhat = model.predict(X)
# retrieve unique clusters
clusters = unique(yhat)

# create scatter plot for samples from each cluster
for cluster in clusters:
    row_ix = where(yhat == cluster) # get row indexes for samples with this cluster
    plt.scatter(X[row_ix, 0], X[row_ix, 1]) # create scatter of these samples

# show the plot
plt.title('Gaussian Mixture Clustering')
plt.show()
```
